Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
# Initialize the map centered at the first launch site
first_site = [launch_sites_df.iloc[0]['Lat'], launch_sites_df.iloc[0]['Long']]
site_map = folium.Map(location=first_site, zoom_start=5)

# Create a MarkerCluster object
marker_cluster = MarkerCluster().add_to(site_map)

# Define marker colors for success (green) and failure (red)
marker_colors = {1: 'green', 0: 'red'}

# Add markers for each launch attempt into the cluster
for index, row in spacex_df.iterrows():
    launch_coordinate = [row['Lat'], row['Long']]
    launch_result = row['class']  # 1 = success, 0 = failure
    
    folium.Marker(
        location=launch_coordinate,
        icon=folium.Icon(color=marker_colors[launch_result], icon="info-sign"),
        popup=f"Launch Site: {row['Launch Site']}<br>Outcome: {'Success' if launch_result == 1 else 'Failure'}"
    ).add_to(marker_cluster)

# Display the map
site_map
```

*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python
# Define a function to assign marker colors based on the class column
def assign_marker_color(launch_class):
    return 'green' if launch_class == 1 else 'red'

# Apply the function to create a new column 'marker_color'
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)

# Display the first few rows to verify
spacex_df.head()

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import radians, cos, sin, sqrt, atan2

# Function to calculate distance between two lat/lon points
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in km
    lat1, lon1, lat2, lon2 = [radians(float(x)) for x in [lat1, lon1, lat2, lon2]]
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c  # Distance in km

# Choose a launch site (e.g., CCAFS LC-40)
selected_site = launch_sites_df[launch_sites_df['Launch Site'] == 'CCAFS LC-40']
site_lat, site_lon = float(selected_site.iloc[0]['Lat']), float(selected_site.iloc[0]['Long'])

# Define proximity locations (example: a city, a highway, and the coastline)
proximity_locations = {
    "Nearest City": (28.3922, -80.6077),  # Cocoa Beach, FL
    "Nearest Highway": (28.5721, -80.6556),  # FL-528 Expressway
    "Coastline": (28.5613, -80.5670)  # Closest ocean point
}

# Calculate distances
distances = {location: haversine(site_lat, site_lon, lat, lon) for location, (lat, lon) in proximity_locations.items()}

# Print results
for location, distance in distances.items():
    print(f"Distance from CCAFS LC-40 to {location}: {distance:.2f} km")

```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
